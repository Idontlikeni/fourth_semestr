Релизовать функцию, оперирующую  векторами длины n.


```python
from math import *


def main(x, y):
    sm = 0
    n = len(x)
    for i in range(1, n+1):
        sm += atan(y[n - ceil(i/2)] - x[ceil(i/4) - 1] ** 3 - 1) ** 7

    return 67 * sm / 66
```

```python
def main(x):
    match x[0]:
        case 2008:
            return x2(x)
        case 1995:
            return x3_1(x)
        case 2001:
            return x3_2(x)


def x2(x):
    match x[2]:
        case 1997:
            return 3
        case 1968:
            return x1_1(x)


def x1_1(x):
    match x[1]:
        case 1990:
            return 0
        case 2018:
            return 1
        case 2007:
            return 2


def x1_2(x, start):
    match x[1]:
        case 1990:
            return 0 + 3 * start
        case 2018:
            return 1 + 3 * start
        case 2007:
            return 2 + 3 * start


def x3_1(x):
    match x[3]:
        case 2019:
            return x1_2(x, 2)
        case 1957:
            return 4
        case 2007:
            return 5


def x3_2(x):
    match x[3]:
        case 1957:
            return x1_2(x, 3)
        case 2019:
            return 13
        case 2007:
            return 12
```

```python
print(main([2001, 2018, 1997, 1957]) == 10,
main([2008, 2018, 1968, 1957]) == 1,
main([2008, 1990, 1968, 1957]) == 0,
main([1995, 1990, 1997, 2019]) == 6,
main([2008, 2018, 1997, 2019]) == 3)
```

```python
def main(n):
    t = bin(int(n, 16))
    t = t[:-1] + '0'
    m = int(t, 2)
    y = []
    arr = [10, 2, 10, 1]
    arr.reverse()
    for i in arr:
        temp = int('0b'+'1' * i, 2)
        y.append(m & temp)
        m = m >> i
    indexes = [2, 3, 0, 1]
    indexes.reverse()
    res = 0
    c = 0
    for i in indexes:
        y[i] = y[i] << c
        res += y[i]
        c += arr[i]
    return hex(res)
```

```python
main('0xe74cd')
```

## Пр 9

```python
def main(s):
    dct = dict()
    s = s.replace("\n", " ")
    s = s.replace("((", "").replace("))", "").replace("<sect>", "")
    s = s.replace("</sect>", "").replace("val", "")
    s = s.replace("#", "")
    s = s.split(".")
    s = list(map(lambda x: x.strip(), s))[:-1]
    #  print(s)
    for i in s:
        t = i.split(' to @\'')
        dct[t[1][:-1]] = int(t[0])
    return dct
```

```python
def main(inp):
    res = sorted(inp, key=lambda x: x[2])
    res = [[i[1], i[2], i[3].split('#')[0], i[3].split('#')[1]] for i in res]
    res = [[*i] for i in list(zip(*res))]
    res[2], res[3] = res[3], res[2]
    for i in range(len(res[0])):
        num = res[0][i][:-1]
        res[0][i] = get_d(num.rjust(2, "0"))
    res[1] = [i[:3] + ' ' + i[4:] for i in res[1]]
    res[3] = [i.replace('@', '[at]') for i in res[3]]
    for i in range(len(res[2])):
        t = res[2][i].split()
        res[2][i] = f'{t[2]}, {t[0][0]}.{t[1]}'
    return res


def get_d(num):
    f = int(num[0])
    if f < 5:
        return f'0.{f + 1 if int(num[1]) > 5 else f}'
    else:
        return str(round(int(num) / 10) / 10)
```
