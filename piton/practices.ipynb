Релизовать функцию, оперирующую  векторами длины n.


```python
from math import *


def main(x, y):
    sm = 0
    n = len(x)
    for i in range(1, n+1):
        sm += atan(y[n - ceil(i/2)] - x[ceil(i/4) - 1] ** 3 - 1) ** 7

    return 67 * sm / 66
```
